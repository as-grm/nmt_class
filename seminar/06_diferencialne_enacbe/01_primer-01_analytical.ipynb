# Izpeljava s pomočjo okolja SageMath

Iščemo rešitev $y=y(x)$ začetnega problema 1. reda

\begin{align}
    y' - y & = 1 - x^2, ~~~ x \in [x_0,x_e]\\
    y(x_0) & = y_0
\end{align}

z začetnim $y(0) = 0.5$.

```python
x = var('x')
y = function('y')(x)
```

```python
de = diff(y,x) == y - x^2 + 1
```

```python
f = desolve(de,y,ics=[0,0.5])
```

```python
display(f)
```

```python
plot(f,x,0,2)
```

```python
plot(f,x,0,4)
```
